2.1)รับค่าตัวเลขตรวจสอบเงื่อนไข

````python
num =input("Enter your number: ").split()
num =list(map(int,num))

print("-------------------------")

for i in num:
    if i < 0:
        sign = "negative"
    elif i > 0:
        sign = "positive"
    else:
        sign = "is zero"

    if i == 0:
        n = "zero"
    elif i % 2 == 0:
        n = "even"
    else:
        n = "is odd"
#challenge
    print(i,"is",sign,"and is",n)

print("-------------------------")


````

```
**ขั้นตอนที่ 1**: ตรวจสอบเงื่อนไขแรกที่ตรรกะ if
โปรแกรมจะพิจารณาประโยคเงื่อนไข:if temp > 20 and not is_raining

โดยยึดตัวเชื่อมหลักคือ and:
  1.ฝั่งซ้าย (temp > 20): แทนค่า 25 > 20 ซึ่งเป็น True
  2.ฝั่งขวา (not is_raining): แทนค่า not True (ตรงข้ามกับ True)
        ผลลัพธ์จึงเปลี่ยนเป็น False
นำผลลัพธ์ทั้งสองฝั่งมาเชื่อมกันด้วย and:
True and False ได้ผลลัพธ์รวมเป็น False

สรุปขั้นตอนนี้: เนื่องจากเงื่อนไขของ if เป็นเท็จ (False) โปรแกรมจึงข้ามคำสั่ง print("Perfect day for a picnic!") ไป และย้ายไปตรวจสอบเงื่อนไขถัดไปแทน



**ขั้นตอนที่ 2**: ตรวจสอบเงื่อนไขถัดมาที่ตรรกะ elif
โปรแกรมจะพิจารณาประโยคเงื่อนไข:elif temp <= 20 or is_raining:
เราแยกคำนวณออกเป็น 2 ฝั่งซ้าย-ขวา

โดยยึดตัวเชื่อมหลักคือ or:
ฝั่งซ้าย (temp <= 20): แทนค่า 25 <= 20 ซึ่งเป็น False
ฝั่งขวา (is_raining): แทนค่าแล้วได้ True ตรงๆนำผลลัพธ์ทั้งสองฝั่งมาเชื่อมกันด้วย or
(เงื่อนไขของ or ขอแค่มีฝั่งใดฝั่งหนึ่งเป็นจริง ผลลัพธ์จะเป็นจริงทันที):False or True  ได้ผลลัพธ์รวมเป็น True

สรุปขั้นตอนนี้: เนื่องจากเงื่อนไขของ elif เป็นจริง (True) โปรแกรมจึงตัดสินใจทำงานในบล็อกนี้ทันที



**ขั้นตอนที่ 3**: การทำงานในบล็อกและข้ามส่วนที่เหลือ
เมื่อเงื่อนไข elif เป็นจริงแล้ว โปรแกรมจะทำคำสั่งด้านในนั่นคือ:
print("Stay indoors or bring an umbrella.")
```
